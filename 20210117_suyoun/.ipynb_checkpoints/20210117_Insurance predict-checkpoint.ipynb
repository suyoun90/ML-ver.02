### Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

```python
x = data.drop(['charges'], axis = 1)
y = data.charges

x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 0)
lr = LinearRegression().fit(x_train,y_train)

y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)

print(lr.intercept_)
print(lr.coef_)
print(lr.score(x_test, y_test))
```

### Ridge & Lasso

모델을 만들 때 bias와 variance를 둘 다 줄여주면 가장 좋겠지만 하나를 포기해야 함.
이 때 bias를 조금 가지더라도 variance를 가장 작게 만들어주면 overfitting을 해결할 수 있지 않을까? 에서 시작.

##### overfitting을 해결하는 방법은 크게 두 가지
1. feature 개수 줄이기
2. 정규화(regularization) 

<정규화> 컨셉을 처음 도입한 모델이 Ridge Regression

##### Ridge (L2 규제)
- 가중치의 절댓값을 가능한한 작게
- 가중치의 모든 원소가 '0에 가깝게'.  but, 정말 0이 되진 않는다.
(모든 특성이 출력에 주는 영향을 최소한으로 만듦) 
- alpha값에 따라 모델을 얼마나 단순화할지 정함. alpha 값을 높이면 계수를 0에 점점 더 가깝게 만들어 성능은 나빠질 수 있으나 일반화에는 도움을 줄 수 있음 

##### Lasso (L1 규제) 
- Ridge와 같이 계수를 0에 가깝게 하지만 어떤 계수는 정말로 0이 됨
- 가중치들이 정말로 0이 되게 함으로써 그로 인한 특성을 제외해 줌. 
- 결과적으로 모델에서 가장 중요한 특성이 무엇인지 알 수 있는 등 모델 해석력이 좋아짐


```python
from sklearn.linear_model import Ridge
Ridge = Ridge(alpha=0.5)
Ridge.fit(x_train, y_train)
print(Ridge.intercept_)
print(Ridge.coef_)
print(Ridge.score(x_test, y_test))
```

```python
from sklearn.linear_model import Lasso
Lasso = Lasso(alpha=0.2, fit_intercept=True, normalize=False, precompute=False, max_iter=1000,
              tol=0.0001, warm_start=False, positive=False, random_state=None, selection='cyclic')
Lasso.fit(x_train, y_train)
print(Lasso.intercept_)
print(Lasso.coef_)
print(Lasso.score(x_test, y_test))
```

## <다항 회귀> 
### PolynomialFeatures() 를 이용한 다항 회귀 

- 다항차수 변환 및 교호작용 변수 생성 : 회귀분석을 할 때 다항 차수를 이용하여 비선형 패턴 (non-linear relation)을 나타내거나 변수간의 곱을 사용해서 interaction effects (교호작용)을 나타낼 수 있는 변수를 만듦
- polynomialFeatures (degree = 2) 를 통해 2차항의 다항 회귀를 만들 수 있음
- ex) 변수 a,b -> ab, a^2, b^2 의 변수 생성
- degree 숫자를 높일수록 추가 되는 조합의 수가 기하급수적으로 늘어나므로 주의! 
- 다항회귀도 선형회귀! 회귀에서 선형/비선형을 나누는 기준은 회귀 계수가 선형/비선형인지에 따른 것이지 독립변수의 선형/비선형 여부와는 무관
- 사이킷런에서는 다항회귀를 위한 클래스가 따로 없지만 다항회귀도 선형회귀이기 때문에 비선형 함수를 선형 모델에 적용시키는 방법을 사용하여 구현

```python
X = data.drop(['charges','region'], axis = 1)
Y = data.charges
quad = PolynomialFeatures (degree = 2)
x_quad = quad.fit_transform(X)
X_train,X_test,Y_train,Y_test = train_test_split(x_quad,Y, random_state = 0)

plr = LinearRegression().fit(X_train,Y_train)
Y_train_pred = plr.predict(X_train)
Y_test_pred = plr.predict(X_test)

print(plr.intercept_)
print(plr.coef_)
print(plr.score(X_test, Y_test))
```
